```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open("./dataset/breast-cancer.labels", "r") as f:
    columns = []
    lines = f.readlines()
    for line in lines:
        columns.append(line.strip("\n"))
    f.close()
cancer_train = pd.io.parsers.read_csv("./dataset/breast-cancer-train.dat", names=columns)
cancer_validate = pd.io.parsers.read_csv("./dataset/breast-cancer-validate.dat", names=columns)
```

```python
plt.hist(cancer_train["symmetry (mean)"])
plt.xlabel("Średnia symetria")
plt.ylabel("Ilość pacjentów")
plt.title("Rozkład średniej symetrii wśród pacjentów")
plt.grid()
plt.show()
```
